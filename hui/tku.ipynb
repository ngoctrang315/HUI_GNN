```python
class RedBlackTree:
    BLACK = True
    RED = False

    class RedBlackTree:
        BLACK = True
    RED = False

    class Node:
        def __init__(self, key=None, color=None, left=None, right=None, parent=None):
            self.key = key
            self.color = color if color is not None else RedBlackTree.BLACK
            self.left = left
            self.right = right
            self.parent = parent

    def __init__(self, allowSameElementMultipleTimes=True):
        self.NULL = self.Node()
        self.size = 0
        self.root = self.NULL
        self.allowSameElementMultipleTimes = allowSameElementMultipleTimes


    def size(self):
        return self.size

    def isEmpty(self):
        return self.root == self.NULL

    def leftRotate(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.NULL:
            y.left.parent = x
        y.parent = x.parent
        if x.parent == self.NULL:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    def rightRotate(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.NULL:
            y.right.parent = x
        y.parent = x.parent
        if x.parent == self.NULL:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y

    def add(self, element):
        z = self.Node(key=element)
        y = self.NULL
        x = self.root

        while x != self.NULL:
            y = x
            compare = z.key.__lt__(x.key)
            if compare:
                x = x.left
            else:
                if z.key == x.key and not self.allowSameElementMultipleTimes:
                    return
                x = x.right
        z.parent = y
        if y == self.NULL:
            self.root = z
        elif z.key.__lt__(y.key):
            y.left = z
        else:
            y.right = z

        self.size += 1
        z.left = self.NULL
        z.right = self.NULL
        z.color = self.RED
        self.insertFixup(z)

    def insertFixup(self, z):
        while z.parent.color == self.RED:
            if z.parent == z.parent.parent.left:
                y = z.parent.parent.right
                if y.color == self.RED:
                    z.parent.color = self.BLACK
                    y.color = self.BLACK
                    z.parent.parent.color = self.RED
                    z = z.parent.parent
                else:
                    if z == z.parent.right:
                        z = z.parent
                        self.leftRotate(z)
                    z.parent.color = self.BLACK
                    z.parent.parent.color = self.RED
                    self.rightRotate(z.parent.parent)
            else:
                y = z.parent.parent.left
                if y.color == self.RED:
                    z.parent.color = self.BLACK
                    y.color = self.BLACK
                    z.parent.parent.color = self.RED
                    z = z.parent.parent
                else:
                    if z == z.parent.left:
                        z = z.parent
                        self.rightRotate(z)
                    z.parent.color = self.BLACK
                    z.parent.parent.color = self.RED
                    self.leftRotate(z.parent.parent)
        self.root.color = self.BLACK

    def transplant(self, u, v):
        if u.parent == self.NULL:
            self.root = v
        elif u == u.parent.left:
            u.parent.left = v
        else:
            u.parent.right = v
        v.parent = u.parent

    def remove(self, element):
        z = self.search(self.root, element)
        if z == self.NULL:
            return
        self.performDelete(z)

    def performDelete(self, z):
        y = z
        yOriginalColor = y.color
        if z.left == self.NULL:
            x = z.right
            self.transplant(z, z.right)
        elif z.right == self.NULL:
            x = z.left
            self.transplant(z, z.left)
        else:
            y = self.minimum(z.right)
            yOriginalColor = y.color
            x = y.right
            if y.parent == z:
                x.parent = y
            else:
                self.transplant(y, y.right)
                y.right = z.right
                y.right.parent = y
            self.transplant(z, y)
            y.left = z.left
            y.left.parent = y
            y.color = z.color
        if yOriginalColor == self.BLACK:
            self.deleteFixup(x)
        self.size -= 1

    def deleteFixup(self, x):
        while x != self.root and x.color == self.BLACK:
            if x == x.parent.left:
                w = x.parent.right
                if w.color == self.RED:
                    w.color = self.BLACK
                    x.parent.color = self.RED
                    self.leftRotate(x.parent)
                    w = x.parent.right
                if w.left.color == self.BLACK and w.right.color == self.BLACK:
                    w.color = self.RED
                    x = x.parent
                else:
                    if w.right.color == self.BLACK:
                        w.left.color = self.BLACK
                        w.color = self.RED
                        self.rightRotate(w)
                        w = x.parent.right
                    w.color = x.parent.color
                    x.parent.color = self.BLACK
                    w.right.color = self.BLACK
                    self.leftRotate(x.parent)
                    x = self.root
            else:
                w = x.parent.left
                if w.color == self.RED:
                    w.color = self.BLACK
                    x.parent.color = self.RED
                    self.rightRotate(x.parent)
                    w = x.parent.left
                if w.right.color == self.BLACK and w.left.color == self.BLACK:
                    w.color = self.RED
                    x = x.parent
                else:
                    if w.left.color == self.BLACK:
                        w.right.color = self.BLACK
                        w.color = self.RED
                        self.leftRotate(w)
                        w = x.parent.left
                    w.color = x.parent.color
                    x.parent.color = self.BLACK
                    w.left.color = self.BLACK
                    self.rightRotate(x.parent)
                    x = self.root
        x.color = self.BLACK

    def search(self, x, k):
        while x != self.NULL and k != x.key:
            if k < x.key:
                x = x.left
            else:
                x = x.right
        return x

    def minimum(self, x):
        while x.left != self.NULL:
            x = x.left
        return x

    def maximum(self, x):
        while x.right != self.NULL:
            x = x.right
        return x

    def lower(self, k):
        result = self.lowerNode(k)
        return result.key if result != self.NULL else None

    def lowerNode(self, k):
        x = self.root
        while x != self.NULL:
            if k > x.key:
                if x.right != self.NULL:
                    x = x.right
                else:
                    return x
            else:
                if x.left != self.NULL:
                    x = x.left
                else:
                    current = x
                    while current.parent != self.NULL and current.parent.left == current:
                        current = current.parent
                    return current.parent
        return self.NULL

    def higher(self, k):
        result = self.higherNode(k)
        return result.key if result != self.NULL else None

    def higherNode(self, k):
        x = self.root
        while x != self.NULL:
            if k < x.key:
                if x.left != self.NULL:
                    x = x.left
                else:
                    return x
            else:
                if x.right != self.NULL:
                    x = x.right
                else:
                    current = x
                    while current.parent != self.NULL and current.parent.right == current:
                        current = current.parent
                    return current.parent
        return self.NULL

    def popMinimum(self):
        if self.root == self.NULL:
            return None
        x = self.root
        while x.left != self.NULL:
            x = x.left
        value = x.key
        self.performDelete(x)
        return value

    def popMaximum(self):
        if self.root == self.NULL:
            return None
        x = self.root
        while x.right != self.NULL:
            x = x.right
        value = x.key
        self.performDelete(x)
        return value

    def contains(self, k):
        return self.search(self.root, k) != self.NULL

    def __str__(self):
        return self._print(self.root, [])

    def _print(self, x, buffer):
        if x != self.NULL and x.key is not None:
            self._print(x.left, buffer)
            buffer.append(str(x.key) + " ")
            self._print(x.right, buffer)
        return "".join(buffer)

    class RedBlackTreeIterator:
        def __init__(self, root):
            self.nodesToVisitNext = []
            self.nodesToVisitNext.append(root)

        def __iter__(self):
            return self

        def __next__(self):
            if not self.nodesToVisitNext:
                raise StopIteration
            currentNode = self.nodesToVisitNext.pop()
            if currentNode.left and currentNode.left.key:
                self.nodesToVisitNext.append(currentNode.left)
            if currentNode.right and currentNode.right.key:
                self.nodesToVisitNext
```

```python
import os
from collections import defaultdict

class AlgoPhase2OfTKU:

    def __init__(self):
        self.minUtility = 0
        self.theCurrentK = 0
        self.numberOfTransactions = 0
        self.inputFilePath = ""
        self.sortedCandidatePath = ""
        self.temporaryFilePathWHUIs = "HUI.txt"
        self.outputTopKHUIsFilePath = ""
        self.delimiter = ":"
        self.numTopKHUI = 0

    def runAlgorithm(self, minUtil, transactionCount, currentK, inputPath, sortedCandidateFile, outputFile):
        self.minUtility = minUtil
        self.numberOfTransactions = transactionCount
        self.theCurrentK = currentK
        self.inputFilePath = inputPath
        self.sortedCandidatePath = sortedCandidateFile
        self.outputTopKHUIsFilePath = outputFile

        with open(self.temporaryFilePathWHUIs, 'w') as bfw:
            HDB = [list() for _ in range(self.numberOfTransactions)]
            BNF = [list() for _ in range(self.numberOfTransactions)]

            # Initialization
            self.initialization(HDB, BNF, len(HDB))

            # Read Database into memory
            self.readDatabase(HDB, BNF, len(HDB), self.inputFilePath)

            # Read Candidate from disk
            self.readCandidateItemsets(HDB, BNF, len(HDB), self.sortedCandidatePath, bfw)

        # Write top-K HUIs to output file
        with open(self.temporaryFilePathWHUIs, 'r') as bfr1, open(self.outputTopKHUIsFilePath, 'w') as bfw1:
            self.numTopKHUI = 0
            for record in bfr1:
                temp = record.split(":")
                if int(temp[1]) >= self.minUtility:
                    # Convert format to SPMF output
                    record = record.replace(":", " #UTIL: ")
                    bfw1.write(record)
                    self.numTopKHUI += 1

        # Delete temporary files
        os.remove(self.temporaryFilePathWHUIs)
        os.remove(self.sortedCandidatePath)

    def readCandidateItemsets(self, HDB, BNF, num_trans, CIPath, Lbfw):
        Heap = RedBlackTree(True)

        with open(CIPath, 'r') as bfr:
            num_HU = 0
            for CIR in bfr:
                CI = CIR.split(self.delimiter)
                Match_Count = 0
                EUtility = 0

                candidate = CI[0].split(" ")

                if int(CI[1]) >= self.minUtility:
                    # For each Candidate, scan DB
                    for i in range(num_trans):
                        if len(HDB[i]) != 0:
                            Match_Count = 0
                            PUtility = 0

                            for s in range(len(candidate)):
                                if int(candidate[s]) in HDB[i]:
                                    Match_Count += 1
                                    index = HDB[i].index(int(candidate[s]))
                                    PUtility += BNF[i][index]
                                else:
                                    PUtility = 0
                                    break

                            if Match_Count == len(candidate):
                                EUtility += PUtility

                    if EUtility >= self.minUtility:
                        Lbfw.write(f"{CI[0]}:{EUtility}\n")
                        self.updateHeap(Heap, CI[0], EUtility)
                        num_HU += 1

        return num_HU

    @staticmethod
    def readDatabase(HDB, BNF, num_trans, DBPath):
        with open(DBPath, 'r') as bfr:
            trans_count = 0
            for record in bfr:
                data = record.split(":")
                transaction = data[0].split(" ")
                benefit = data[2].split(" ")

                for i in range(len(transaction)):
                    HDB[trans_count].append(int(transaction[i]))
                    BNF[trans_count].append(int(benefit[i]))

                trans_count += 1

    @staticmethod
    def initialization(HDB, BNF, num_trans):
        for i in range(num_trans):
            HDB[i] = []
            BNF[i] = []

    def updateHeap(self, NCH, HUI, Utility):
        if NCH.size() < self.theCurrentK:
            NCH.add((HUI, Utility))
        elif NCH.size() >= self.theCurrentK:
            if Utility > self.minUtility:
                NCH.add((HUI, Utility))
                NCH.popMinimum()

        if NCH.minimum()[1] > self.minUtility and NCH.size() >= self.theCurrentK:
            self.minUtility = NCH.minimum()[1]

    def getNumberOfTopKHUIs(self):
        return self.numTopKHUI

    def setNumberOfTopKHUIs(self, numTopKHUI):
        self.numTopKHUI = numTopKHUI
```

```python
import os

class CalculateDatabaseInfo:
    def __init__(self, inputPath):
        self.inputPath = inputPath
        self.totalUtility = 0
        self.databaseSize = 0
        self.maxID = 0
        self.sumAllLength = 0
        self.avgLength = 0.0
        self.maxLength = 0
        self.allItem = set()

    def runCalculate(self):
        try:
            with open(self.inputPath, 'r') as br:
                for line in br:
                    self.databaseSize += 1
                    tokens1 = line.strip().split(":")  # Chia thành 3 phần: itemsets, transaction utility, utility của mỗi item
                    tokens2 = tokens1[0].split(" ")  # Chia itemsets thành các item
                    self.totalUtility += int(tokens1[1])
                    self.sumAllLength += len(tokens2)
                    if self.maxLength < len(tokens2):
                        self.maxLength = len(tokens2)

                    for token in tokens2:
                        num = int(token)
                        if num > self.maxID:
                            self.maxID = num
                        self.allItem.add(num)

            self.avgLength = round(self.sumAllLength / self.databaseSize, 2)  # Làm tròn đến 2 chữ số thập phân

        except Exception as e:
            print(e)
            return False  # failure

        return True

    # Ghi kết quả ra file
    def outputResult(self, outputPath):
        try:
            with open(outputPath, 'w') as output:
                print("----------Database Information----------")
                print(f"Input file path : {self.inputPath}")
                print(f"Output file path : {outputPath}")
                print(f"Number of transactions : {self.databaseSize}")
                print(f"Total utility : {self.totalUtility}")
                print(f"Number of distinct items : {len(self.allItem)}")
                print(f"Maximum Id of item : {self.maxID}")
                print(f"Average length of transaction : {self.avgLength}")
                print(f"Maximum length of transaction : {self.maxLength}")

                output.write("----------Database Information----------\n")
                output.write(f"Input file path : {self.inputPath}\n")
                output.write(f"Output file path : {outputPath}\n")
                output.write(f"Number of transactions : {self.databaseSize}\n")
                output.write(f"Total utility : {self.totalUtility}\n")
                output.write(f"Number of distinct items : {len(self.allItem)}\n")
                output.write(f"Maximum Id of item : {self.maxID}\n")
                output.write(f"Average length of transaction : {self.avgLength}\n")
                output.write(f"Maximum length of transaction : {self.maxLength}\n")

        except Exception as e:
            print(f"Error writing to file: {e}")

    def getMaxID(self):
        return self.maxID

    def getMaxLength(self):
        return self.maxLength

    def getDBSize(self):
        return self.databaseSize
```

```python
import os
import psutil
import sys

class MemoryLogger:
    # Singleton instance
    instance = None

    def __init__(self):
        # Variable to store the maximum memory usage
        self.maxMemory = 0

        # Boolean flag to indicate whether the recording mode is on or off
        self.recordingMode = False

        # File object to store the output file name
        self.outputFile = None

        # Writer object to write the memory usage values to the file
        self.writer = None

    @staticmethod
    def getInstance():
        if MemoryLogger.instance is None:
            MemoryLogger.instance = MemoryLogger()
        return MemoryLogger.instance

    def getMaxMemory(self):
        return self.maxMemory

    def reset(self):
        self.maxMemory = 0

    def checkMemory(self):
        # Get current memory usage in megabytes
        process = psutil.Process(os.getpid())
        currentMemory = process.memory_info().rss / 1024 / 1024

        if currentMemory > self.maxMemory:
            self.maxMemory = currentMemory

        # If recording mode is on, write the current memory usage to the file
        if self.recordingMode and self.writer is not None:
            try:
                self.writer.write(f"{currentMemory}\n")
            except Exception as e:
                print(f"Error writing memory usage to file: {e}")

        return currentMemory

    def startRecordingMode(self, fileName):
        # Set the recording mode flag to True
        self.recordingMode = True
        self.outputFile = fileName
        # Open the file for writing
        try:
            self.writer = open(fileName, 'w')
        except Exception as e:
            print(f"Error opening file for recording: {e}")

    def stopRecordingMode(self):
        # If recording mode is on, close the file
        if self.recordingMode and self.writer is not None:
            try:
                self.writer.close()
            except Exception as e:
                print(f"Error closing file: {e}")
            self.recordingMode = False
```

```python
class TKUTriangularMatrix:
    def __init__(self, elementCount):
        self.elementCount = elementCount
        self.matrix = [[] for _ in range(elementCount)]
        for i in range(elementCount):
            self.matrix[i] = [0] * (elementCount - i)

    def get(self, i, j):
        """ Get the value at the position (i, j) in the matrix. """
        if j < i:
            return self.matrix[j][self.elementCount - i - 1]
        else:
            return self.matrix[i][self.elementCount - j - 1]

    def incrementCount(self, id1, id2, sum_val):
        """ Increment the count in the matrix by some value. """
        if id2 < id1:
            self.matrix[id2][self.elementCount - id1 - 1] += sum_val
        else:
            self.matrix[id1][self.elementCount - id2 - 1] += sum_val

    def getSupportForItems(self, id1, id2):
        """ Get the support for the pair of items (id1, id2) in the matrix. """
        if id2 < id1:
            return self.matrix[id2][self.elementCount - id1 - 1]
        else:
            return self.matrix[id1][self.elementCount - id2 - 1]

    def __str__(self):
        """ Get the string representation of the matrix. """
        result = []
        result.append(f"Element count = {self.elementCount}")
        for i in range(len(self.matrix)):
            result.append(f"{i}: " + " ".join(map(str, self.matrix[i])))
        return "\n".join(result)


# Test the TKUTriangularMatrix class
if __name__ == "__main__":
    a = TKUTriangularMatrix(5)
    
    print(a)
    a.incrementCount(1, 2, 1)
    print("add {1 2}")
    print(a)

    print("add {1 2}")
    a.incrementCount(1, 2, 1)
    print(a)

    print("add {1 3}")
    a.incrementCount(1, 3, 1)
    print(a)
    
    a.incrementCount(1, 4, 1)
    print(a)

    a.incrementCount(1, 3, 1)
    a.incrementCount(2, 4, 1)
    a.incrementCount(2, 4, 1)
    a.incrementCount(4, 3, 1)
    print(a)
    
    a.incrementCount(0, 2, 1)
    a.incrementCount(0, 3, 1)
    a.incrementCount(0, 4, 1)
    print(a)
```

```python
import time
from collections import defaultdict, deque
import heapq

class AlgoTKU:

    def __init__(self):
        self.theInputFile = None
        self.theCandidateFile = None
        self.kValue = None
        self.itemCount = None
        self.globalMinUtil = 0
        self.arrayTWUItems = []
        self.arrayMIU = []
        self.arrayMAU = []
        self.totalTime = 0
        self.patternCount = 0

    def runAlgorithm(self, inputFile, outputFile, k):
        self.totalTime = time.time()
        self.globalMinUtil = 0
        self.kValue = k
        self.theInputFile = inputFile
        self.theCandidateFile = "topKcandidate.txt"

        # Initialize and calculate database info
        tool = CalculateDatabaseInfo(inputFile)
        tool.runCalculate()
        self.itemCount = tool.getMaxID() + 1

        self.arrayTWUItems = [0] * self.itemCount
        self.arrayMIU = [0] * self.itemCount
        self.arrayMAU = [0] * self.itemCount

        with open(self.theCandidateFile, 'w') as bfw_CI:
            # Phase 1: Generate P1 and pre-evaluation
            self.globalMinUtil = self.preEvaluation(self.theInputFile, self.arrayTWUItems, self.itemCount, self.arrayMIU, self.arrayMAU, self.globalMinUtil, self.kValue)

            # Build FP-Tree
            tree = self.BuildUPTree(self.arrayTWUItems, self.theInputFile)

            # Sum Descendents
            DSNodeCountHeap = []

            for child in tree.root.childlink:
                Sum_DS = [0] * self.itemCount
                DSItem = child.item
                tree.SumDescendent(child, Sum_DS)

                for j in range(len(Sum_DS)):
                    if Sum_DS[j] != 0 and j != DSItem:
                        DS_Value = (self.arrayMIU[j] + self.arrayMIU[DSItem]) * Sum_DS[j]
                        self.UpdateNodeCountHeap(DSNodeCountHeap, DS_Value)

            DSNodeCountHeap = []

            # Get Ulist
            ulist = []
            self.getUlist(self.arrayTWUItems, ulist)

            # Mining phase
            prefix = ""
            tree.UPGrowth(tree, ulist, prefix, bfw_CI, [], self.arrayTWUItems)

        # Phase 2: Sort HUI
        sortedTopKcandidateFile = "sortedTopKcandidate.txt"
        self.runSortHUIAlgorithm(self.theCandidateFile, sortedTopKcandidateFile)

        # Delete temporary file
        import os
        os.remove(self.theCandidateFile)

        # Phase 3
        algoPhase2 = AlgoPhase2OfTKU()
        algoPhase2.runAlgorithm(int(self.globalMinUtil), tool.getDBSize(), k, inputFile, sortedTopKcandidateFile, outputFile)
        self.patternCount = algoPhase2.getNumberOfTopKHUIs()

        # Total execution time
        self.totalTime = time.time() - self.totalTime

    def runSortHUIAlgorithm(self, theCandidateFile, sortedTopKcandidateFile):
        Heap = []

        with open(theCandidateFile, 'r') as bfr:
            for record in bfr:
                temp = record.strip().split(":")
                Heap.append((-int(temp[1]), temp[0]))  # Negate for max-heap

        heapq.heapify(Heap)

        with open(sortedTopKcandidateFile, 'w') as bfw1:
            while Heap:
                max_val = heapq.heappop(Heap)
                bfw1.write(f"{max_val[1]}:{-max_val[0]}\n")

    def preEvaluation(self, HDB, TWU1, num_Item, MinBNF, MaxBNF, mini_utility, pK):
        a = TKUTriangularMatrix(num_Item)

        with open(HDB, 'r') as bfr:
            for transaction in bfr:
                temp1 = transaction.split(":")
                temp2 = temp1[0].split(" ")
                temp3 = temp1[2].split(" ")

                for s in range(len(temp2)):
                    MinBNF[int(temp2[s])] = min(MinBNF[int(temp2[s])], int(temp3[s])) if MinBNF[int(temp2[s])] > 0 else int(temp3[s])
                    MaxBNF[int(temp2[s])] = max(MaxBNF[int(temp2[s])], int(temp3[s]))
                    TWU1[int(temp2[s])] += int(temp1[1])

                    if s > 0:
                        a.incrementCount(int(temp2[0]), int(temp2[s]), int(temp3[0]) + int(temp3[s]))

        return self.getInitialUtility(a, num_Item, pK)

    def getInitialUtility(self, TM, nItem, K):
        topKList = []

        for i in range(nItem):
            for j in range(len(TM.matrix[i])):
                if TM.matrix[i][j] != 0:
                    if len(topKList) < K:
                        heapq.heappush(topKList, TM.matrix[i][j])
                    elif TM.matrix[i][j] > topKList[0]:
                        heapq.heappushpop(topKList, TM.matrix[i][j])

        return heapq.heappop(topKList)

    def getUlist(self, P1, list):
        for i in range(len(P1)):
            if P1[i] > 0 and P1[i] >= self.globalMinUtil:
                self.InsertItem(list, i, P1)

    def InsertItem(self, list, target, Order):
        for i in range(len(list)):
            if Order[target] > Order[list[i]]:
                list.insert(i, target)
                return
            elif Order[target] == Order[list[i]] and target < list[i]:
                list.insert(i, target)
                return
        list.append(target)

    def UpdateNodeCountHeap(self, NCH, NewValue):
        if len(NCH) < self.kValue:
            heapq.heappush(NCH, NewValue)
        elif NewValue > self.globalMinUtil:
            heapq.heappushpop(NCH, NewValue)

        if NCH[0] > self.globalMinUtil and len(NCH) >= self.kValue:
            self.globalMinUtil = NCH[0]

    def BuildUPTree(self, P1, HDB):
        tree = FPtree(self.itemCount)
        with open(HDB, 'r') as bfr:
            for transaction in bfr:
                temp1 = transaction.split(":")
                temp2 = temp1[0].split(" ")
                bran = temp1[2].split(" ")
                bran2 = []
                tran = []
                tranlen = 0

                for m in range(len(temp2)):
                    if P1[int(temp2[m])] >= self.globalMinUtil:
                        bran2.append(bran[m])
                        tran.append(int(temp2[m]))
                        tranlen += 1

                self.sorttrans2(tran, bran2, 0, tranlen, P1)
                tree.instrans3(tran, bran2, tranlen, P1, 1, [])

        return tree

    def sorttrans2(self, tran, bran, pre, tranlen, P1):
        for i in range(pre, tranlen - 1):
            for j in range(pre, tranlen - 1):
                if P1[tran[j]] < P1[tran[j + 1]]:
                    tran[j], tran[j + 1] = tran[j + 1], tran[j]
                    bran[j], bran[j + 1] = bran[j + 1], bran[j]
                elif P1[tran[j]] == P1[tran[j + 1]] and tran[j] > tran[j + 1]:
                    tran[j], tran[j + 1] = tran[j + 1], tran[j]
                    bran[j], bran[j + 1] = bran[j + 1], bran[j]
```
