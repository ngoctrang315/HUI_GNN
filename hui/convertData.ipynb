```python
import pandas as pd

# Giả sử dữ liệu được đọc vào từ file CSV, ở đây ta tạo dataframe mẫu từ dữ liệu bạn cung cấp
data = pd.read_csv("Sales.csv")

df = pd.DataFrame(data)

# Tạo ra dữ liệu có cấu trúc HUI: id các mặt hàng, tiện ích tổng cộng và tiện ích chi tiết
hui_data = []

# Nhóm dữ liệu theo Order Num và Line Item để tính tiện ích
for order_num, group in df.groupby('OrderNumber'):
    items = group['ProductKey'].tolist()
    total_utility = group['Quantity'].sum()
    item_utilities = group['Quantity'].tolist()
    
    hui_entry = f"{' '.join(map(str, items))}:{total_utility}:{' '.join(map(str, item_utilities))}"
    hui_data.append(hui_entry)

# Kết quả
for entry in hui_data:
    print(entry)
```

```python
import pandas as pd

# Dữ liệu Product.csv
product_data = pd.read_csv("Products.csv")

df_product = pd.DataFrame(product_data)

# Giả sử dữ liệu từ Sales.csv đã được đọc vào và có cấu trúc như sau
sales_data = pd.read_csv("Sales.csv")

df_sales = pd.DataFrame(sales_data)

# Hàm tính lợi nhuận cho mỗi sản phẩm
def calculate_profit(product_key, quantity):
    product = df_product[df_product['ProductKey'] == product_key]
    if not product.empty:
        unit_profit = product['UnitPriceUSD'].values[0] - product['UnitCostUSD'].values[0]
        return unit_profit * quantity
    else:
        return 0

# Tạo ra dữ liệu DB_Utility (ProductKey : Tổng lợi nhuận : Lợi nhuận của từng ProductKey)
db_utility_data = []

for order_num, group in df_sales.groupby('OrderNumber'):
    product_keys = group['ProductKey'].tolist()
    quantities = group['Quantity'].tolist()
    
    # Tính lợi nhuận cho từng ProductKey
    profits = [calculate_profit(product_key, quantity) for product_key, quantity in zip(product_keys, quantities)]
    
    # Tính tổng lợi nhuận cho cả giao dịch
    total_profit = sum(profits)
    
    # Tạo chuỗi DB_Utility theo định dạng yêu cầu
    utility_entry = f"{' '.join(map(str, product_keys))}:{total_profit:.2f}:{' '.join(map(lambda x: f'{x:.2f}', profits))}"
    db_utility_data.append(utility_entry)

# Kết quả DB_Utility
db_utility_data
```

```python
import pandas as pd

# Dữ liệu Product.csv
product_data = pd.read_csv("Products.csv")

df_product = pd.DataFrame(product_data)

# Giả sử dữ liệu từ Sales.csv đã được đọc vào và có cấu trúc như sau
sales_data = pd.read_csv("Sales.csv")

df_sales = pd.DataFrame(sales_data)

# Hàm tính lợi nhuận cho mỗi sản phẩm
def calculate_profit(product_key, quantity):
    product = df_product[df_product['ProductKey'] == product_key]
    if not product.empty:
        unit_profit = product['UnitPriceUSD'].values[0] - product['UnitCostUSD'].values[0]
        return unit_profit * quantity
    else:
        return 0

# Lọc các transaction có từ 3 item trở lên
db_utility_data_filtered = []

for order_num, group in df_sales.groupby('OrderNumber'):
    product_keys = group['ProductKey'].tolist()
    if len(product_keys) >= 3:  # Chỉ lấy các giao dịch có từ 3 sản phẩm trở lên
        quantities = group['Quantity'].tolist()
        profits = [round(calculate_profit(product_key, quantity)) for product_key, quantity in zip(product_keys, quantities)]
        total_profit = round(sum(profits))
        utility_entry = f"{' '.join(map(str, product_keys))}:{total_profit}:{' '.join(map(str, profits))}"
        db_utility_data_filtered.append(utility_entry)

# Lưu tất cả các giao dịch đã lọc vào file txt
file_all_filtered_path = "db_utility_data_filtered_3.txt"
with open(file_all_filtered_path, "w") as file:
    for entry in db_utility_data_filtered:
        file.write(entry + "\n")

# Lưu những ProductKey tìm được từ db_utility_data_filtered
file_product_key_filtered_path = "db_utility_productkey_filtered_3.txt"
product_keys_filtered = set()

with open(file_product_key_filtered_path, "w") as file:
    for entry in db_utility_data_filtered:
        product_keys = entry.split(":")[0]  # Lấy phần ProductKey
        product_keys_filtered.update(map(int, product_keys.split()))  # Lưu tất cả ProductKey vào set để loại trùng lặp
        file.write(product_keys + "\n")

# Lưu những ProductName tương ứng với ProductKey đã lọc
file_product_name_filtered_path = "db_utility_productname_filtered_3.txt"

with open(file_product_name_filtered_path, "w") as file:
    for entry in db_utility_data_filtered:
        product_keys = list(map(int, entry.split(":")[0].split()))  # Lấy ProductKey từ chuỗi
        product_names = [df_product[df_product['ProductKey'] == key]['ProductName'].values[0] for key in product_keys]
        product_name_entry = f"{' '.join(map(str, product_keys))}: {'; '.join(product_names)}"
        file.write(product_name_entry + "\n")

# Trả về đường dẫn của các file
file_all_filtered_path, file_product_key_filtered_path, file_product_name_filtered_path
```
